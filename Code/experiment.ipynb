```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
import pandas as pd
```

```python
# Load datasets
train_orig = pd.read_csv("train_original.csv")

test_random = pd.read_csv("test_random.csv")
test_random_scaled = pd.read_csv("test_random_scaled.csv")

test_near = pd.read_csv("test_near.csv")
test_near_scaled = pd.read_csv("test_near_scaled.csv")
```

```python
# Train model on original data
rf_orig = RandomForestClassifier(random_state=42, n_estimators=100)
rf_orig.fit(train_orig[["X1", "X2"]], train_orig["y"])
```

```python
# Evaluate original model
acc_orig_random = accuracy_score(test_random["y"], rf_orig.predict(test_random[["X1", "X2"]]))
acc_orig_near   = accuracy_score(test_near["y"], rf_orig.predict(test_near[["X1", "X2"]]))

# Evaluate scaled model
acc_scaled_random = accuracy_score(test_random_scaled["y"], rf_orig.predict(test_random_scaled[["X1", "X2"]]))
acc_scaled_near   = accuracy_score(test_near_scaled["y"], rf_orig.predict(test_near_scaled[["X1", "X2"]]))
```

```python
# Print results
print(f"Train = original | Test = random          â†’ Accuracy: {acc_orig_random:.3f}")
print(f"Train = original | Test = near-split      â†’ Accuracy: {acc_orig_near:.3f}")
print(f"Train = original | Test = random_scaled   â†’ Accuracy: {acc_scaled_random:.3f}")
print(f"Train = original | Test = near_scaled     â†’ Accuracy: {acc_scaled_near:.3f}")
```
